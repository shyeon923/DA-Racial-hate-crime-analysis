```python
path_5=os.path.join("/Users/mandyhong/Desktop/DA401", "state_abbr.csv")
stateL=pd.read_csv(path_5, names=["name", "abbr"])
stateL
```

```python
stateD=stateL.to_dict(orient="list")

newD={}
#newD[stateD['name'][1]]=stateD['abbr'][1]
#newD

#newD[stateD['name'][2]]=stateD['abbr'][2]
#newD

for i in range(0,len(stateD['name'])):
    #print(i)
    newD[stateD['name'][i]]=stateD['abbr'][i]


newD
```

```python
#1. first, add state abbrevation column to the census data
#change string type month into integer month
stateD = newD
census["STATE_ABBR"] = census.STATE.map(stateD)
census
```

```python
hc=hc.rename({"state":"STATE_ABBR", "year":"YEAR", "hate_crime":"ALL_HATE_CRIME"}, axis=1)
black_hc=black_hc.rename({"state":"STATE_ABBR", "year":"YEAR", "hate_crime":"BLACK_HATE_CRIME"}, axis=1)
asian_hc=asian_hc.rename({"state":"STATE_ABBR", "year":"YEAR", "hate_crime":"ASIAN_HATE_CRIME"}, axis=1)
asian_hc
```

```python
#hc.loc[hc["STATE_ABBR"]=="NE"]
hc[hc['STATE_ABBR'].str.contains("N")] #Problem with wrond STATE_ABBR from FBI data. NB changed to NE
hc['STATE_ABBR'] = hc['STATE_ABBR'].replace(['NB'], 'NE')
hc[hc['STATE_ABBR'].str.contains("N")]

#same problem with these two data as well
black_hc.loc[black_hc["STATE_ABBR"]=="NB"]
asian_hc.loc[asian_hc["STATE_ABBR"]=="NB"]
black_hc['STATE_ABBR'] = black_hc['STATE_ABBR'].replace(['NB'], 'NE')
asian_hc['STATE_ABBR'] = asian_hc['STATE_ABBR'].replace(['NB'], 'NE')

asian_hc.loc[asian_hc["STATE_ABBR"]=="NE"]
```

```python
#2. change column name for long column names
census = census.rename({"year":"YEAR","Total!!Estimate!!Population 18 to 24 years!!Bachelor's degree or higher": '18_24_BACHELOR_HIGHER', "Total!!Estimate!!Population 25 to 34 years!!Bachelor's degree or higher": '25_34_BACHELOR_HIGHER', "Total!!Estimate!!Population 35 to 44 years!!Bachelor's degree or higher": "35_44_BACHELOR_HIGHER", "Total!!Estimate!!Population 45 to 64 years!!Bachelor's degree or higher":"45_64_BACHELOR_HIGHER", "Total!!Estimate!!Population 65 years and over!!Bachelor's degree or higher":"65_MORE_BACHELOR_HIGHER"}, axis='columns')
census
```

```python
census["WHITE_PER_CAPITA"]=census["WHITE"]/census["POP"]
census["BLACK_PER_CAPITA"]=census["BLACK"]/census["POP"]
census["ASIAN_PER_CAPITA"]=census["ASIAN"]/census["POP"]
census["ONE_RACE_RATIO"]=census["ONE_RACE_POP"]/census["POP"]
census["FOREIGN_BORN_NON_US_RATIO"]=census["FOREIGN_BORN_NON_US"]/census["POP"]
census["SCHOOL_ENROLLMENT_RATE"]=census["SCHOOL_ENROLLMENT"]/census["POP"]
census["BACHELOR_HIGHER_RATIO"]=None
census
```

```python
census.loc[0,"18_24_BACHELOR_HIGHER":"65_MORE_BACHELOR_HIGHER"].sum()/census.loc[0,"POP"]

for i in range(0,len(census)):
    census["BACHELOR_HIGHER_RATIO"][i]=census.loc[i,"18_24_BACHELOR_HIGHER":"65_MORE_BACHELOR_HIGHER"].sum()/census.loc[i,"POP"]
census
```

```python
census["STATE"][0]
census=census.sort_values(by=['STATE', 'YEAR'])
census
```

```python
new_census=pd.merge(census, hc, left_on=['STATE_ABBR','YEAR'], right_on=["STATE_ABBR","YEAR"], how='left')
new_census=pd.merge(new_census, black_hc, left_on=['STATE_ABBR','YEAR'], right_on=["STATE_ABBR","YEAR"], how='left')
new_census=pd.merge(new_census, asian_hc, left_on=['STATE_ABBR','YEAR'], right_on=["STATE_ABBR","YEAR"], how='left')

#new_census=new_census.drop(columns=[""])
CENSUS=new_census.copy()
```

```python
CENSUS["ALL_HC_PER_CAPITA"]=CENSUS["ALL_HATE_CRIME"]/census["POP"]*100000 #x number of hate crime victim per every 100,000 people in the state.
CENSUS["BLACK_HC_PER_CAPITA_POP"]=CENSUS["BLACK_HATE_CRIME"]/census["POP"]*100000
CENSUS["BLACK_HC_PER_CAPITA_RACE_POP"]=CENSUS["BLACK_HATE_CRIME"]/census["BLACK"]*100000
CENSUS["ASIAN_HC_PER_CAPITA_POP"]=CENSUS["ASIAN_HATE_CRIME"]/census["POP"]*100000
CENSUS["ASIAN_HC_PER_CAPITA_RACE_POP"]=CENSUS["ASIAN_HATE_CRIME"]/census["ASIAN"]*100000
CENSUS
```

```python
CENSUS["WHITE_PER_CAPITA"]=CENSUS["WHITE_PER_CAPITA"]*100
CENSUS["BLACK_PER_CAPITA"]=CENSUS["BLACK_PER_CAPITA"]*100
CENSUS["ASIAN_PER_CAPITA"]=CENSUS["ASIAN_PER_CAPITA"]*100
CENSUS["ONE_RACE_RATIO"]=CENSUS["ONE_RACE_RATIO"]*100
CENSUS["FOREIGN_BORN_NON_US_RATIO"]=CENSUS["FOREIGN_BORN_NON_US_RATIO"]*100
CENSUS["SCHOOL_ENROLLMENT_RATE"]=CENSUS["SCHOOL_ENROLLMENT_RATE"]*100
CENSUS["BACHELOR_HIGHER_RATIO"]=CENSUS["BACHELOR_HIGHER_RATIO"]*100
```
